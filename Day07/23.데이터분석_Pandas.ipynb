```python
# Fare -> LowFare, MediumFare, HighFare
# LowFare = 15, MediumFare = 30 , HighFare > 30
train['LowFare'] = train.Fare < 15
train['MediumFare'] = (train.Fare >= 15) & (train.Fare < 30)
train['HighFare'] = train.Fare >= 30

train[['Fare', 'LowFare', 'MediumFare', 'HighFare']].head(10)
```

> One Hot Encoding    
수치의 크기가 의미없는 컬럼에 대해 True와 False로 컬럼을 구분하여 저장한다.

```python
# FareType이라는 컬럼에
# 15보다 작으면 Low
# 15 ~ 30 이면 Medium
# 30초과면 High
train.loc[train.Fare < 15, "FareType"] = "Low"
train.loc[(train.Fare >= 15) & (train.Fare < 30), "FareType"] = "Medium"
train.loc[train.Fare >= 30, "FareType"] = "High"

train[['Fare', 'FareType']].head(10)
```

### Python의 함수를 이용하여 처리하기

```python
def check_fare_type(fare):
  return_value = ""
  if fare < 15:
    return_value = "Low"
  elif fare >= 15 and fare < 30:
    return_value = "Medium"
  else:
    return_value = "High"
    
  return return_value
```

```python
train['FareType2'] = train.Fare.apply(check_fare_type)
train[['Fare','FareType','FareType2']].head(10)
```
