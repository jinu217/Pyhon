### 지도에 마커 표시하기
- 서울 시내 주요 대학교의 위치를 표시하기

```python
# 대학교 정보 가져오기
df = pd.read_excel("./Data/서울지역 대학교 위치.xlsx")
df.columns=['학교명','위도','경도']
df.head()
```

```python
df.info()
```

```python
# 서울지도 만들기
seoul_map = folium.Map(
  location=[37.55, 126.98],
  zoom_start=11
)

# 대학교 위치정보를 Marker로 표시

for name, lat, lng in zip(df.학교명, df.위도, df.경도):
  popup=folium.Popup(name, max_width=200) # pop up의 글자를 가로로 배치 
  folium.Marker(
    [lat, lng],
    popup=popup,
  ).add_to(seoul_map)

seoul_map
```

---
### 지도에 원형 마커 표시

```python
# 서울지도 만들기
seoul_map = folium.Map(
  location=[37.55, 126.98],
  zoom_start=11
)

# 대학교 위치정보를 Marker로 표시

for name, lat, lng in zip(df.학교명, df.위도, df.경도):
  folium.CircleMarker(
    [lat, lng],
    radius=10,  # 원의 반지름
    color='brown',  # 원의 둘레 색상
    fill=True,
    fill_color='coral', #원을 채우는 색
    fill_opacity=0.7,
  ).add_to(seoul_map)

seoul_map
```

### 지도 영역에 단계 구분표(Choropleth Map) 표시하기
- 행정구역과 같이 지도상의 어떤 경계에 둘러싸인 영역에 색을 칠하거나 음영등의 정보를 나타내는 시각화 기법

```python
import json
```

```python
# 데이터 불러오기
df = pd.read_excel("./Data/경기도인구데이터.xlsx", index_col='구분')
df.head()
```

```python
# 경기도 시군구 경계 파일 불러오기
geo_path = "./Data/경기도행정구역경계.json"
geo_data = json.load(
  open(geo_path, encoding='utf-8')
)

geo_data
```

```python
# 경기도 지도 만들기
g_map = folium.Map(
  location=[37.55, 126.98],
  zoom_start=9
)

# 출력할 연도 선택
year = 2015

# Choropleth 표시하기
folium.Choropleth(
  geo_data=geo_data, # 지도 경계
  data=df[year],
  columns=[df.index, df[year]],
  fill_color='YlOrRd',
  threshold_scale=[10000, 100000, 300000, 500000, 700000],
  key_on='feature.properties.name'
).add_to(g_map)

g_map
```
