## 서울시 교통사고 시각화

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import folium 
```

```python
traffic_seoul = pd.read_csv("./Data/newSeoul_2005_2019.csv")
traffic_seoul.head()
```

```python
traffic_seoul.tail()
```

```python
traffic_seoul.info()
```

### 컬럼별 상관계수 구하기

```python
np.corrcoef(traffic_seoul['발생건수'], traffic_seoul['부상자수'])
```

```python
np.corrcoef(traffic_seoul['발생건수'], traffic_seoul['사망자수'])
```

```python
np.corrcoef(traffic_seoul['부상자수'], traffic_seoul['사망자수'])
```

```python
# 2019년 데이터만 추출
traffic_seoul_2019 = \
  traffic_seoul[traffic_seoul['년도'] == 2019]

traffic_seoul_2019
```

```python
# index 정리하기
traffic_seoul_2019.reset_index(drop=True, inplace=True)
traffic_seoul_2019
```

```python
# 자치구별 발생건수, 사망자수, 부상자수 집계하기
trffic_anal = \
  pd.pivot_table(
    traffic_seoul_2019,
    index='자치구명',
    aggfunc=sum
  )

trffic_anal
```

```python
# 년도와 월은 제거
trffic_anal = \
  trffic_anal[['발생건수','부상자수','사망자수']]

trffic_anal.head()
```

```python
# 위도 경도를 포함한 데이터
seoul_limit = pd.read_csv("./Data/seoul.csv", encoding='euc-kr')
seoul_limit.columns = ['자치구명', 'lon', 'lat']
seoul_limit.head()
```

```python
# traffic_anal과 seoul_limit 병합하기
data_result = pd.merge(trffic_anal, seoul_limit, on='자치구명')
data_result.head()
```

```python
data_result.info()
```

```python
# folium을 사용하기 위해 int를 float로 변환해야 한다.

data_result['발생건수'] = data_result['발생건수'].astype(float)
data_result['부상자수'] = data_result['부상자수'].astype(float)
data_result['사망자수'] = data_result['사망자수'].astype(float)

data_result.info()
```

```python
# 자치구청 표시
map = folium.Map(
  location=[37.5502, 126.982],
  zoom_start=11
)

for n in data_result.index:
  popup = folium.Popup(
    data_result['자치구명'][n],
    max_width=200
  )
  folium.Marker(
    [data_result['lat'][n], data_result['lon'][n]],
    popup=popup
  ).add_to(map)

map
```

### 자치구별 교통사고 발생건수 표시하기

```python
data_result.head()
```

```python
# 자치구청 표시
map = folium.Map(
  location=[37.5502, 126.982],
  zoom_start=11
)

for n in data_result.index:
  popup = folium.Popup(
    data_result['자치구명'][n],
    max_width=200
  )
  folium.Marker(
    [data_result['lat'][n], data_result['lon'][n]],
    popup=popup
  ).add_to(map)

  folium.CircleMarker(
    [data_result['lat'][n], data_result['lon'][n]],
    radius=data_result['발생건수'][n] / 100,
    color = 'blue',
    fill_color = 'skyblue',
    fill = True,
    alpha = 0.5
  ).add_to(map)

map
```
