# Matplotlib

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("./Data/시도별 전출입 인구수.xlsx", header=0)
df.head()
```

```python
df.tail()
```

```python
df.info()
```

```python
df.head()
```

```python
# NaN을 전(Prev) 데이터로 채움 (엑셀 양식 병합 자료 처리)
df = df.ffill()
df.head()
```

```python
df.tail()
```

```python
# 서울에서 다른 지역으로 이동한 데이터만 추출
mask = (df.전출지별 == '서울특별시') & (df.전입지별 != '서울특별시')
mask
```

```python
df_seoul = df[mask]
```

```python
# 전출지별 컬럼 삭제
df_seoul.drop('전출지별', axis='columns', inplace=True)
df_seoul.head()
```

```python
# 전입지별 컬럼 이름을 전입지로 변경
df_seoul.rename(
  {'전입지별' : '전입지'},
  axis='columns',
  inplace=True
)
df_seoul.head()
```

```python
# df_seoul의 index를 전입지 컬럼으로 변경
df_seoul.set_index('전입지', inplace=True)
df_seoul.head()
```

```python
# 서울에서 경기도로 이동한 인구 데이터 값만 선택

sr_one = df_seoul.loc['경기도',:]
sr_one
```

```python
# x,y축 데이터를 정한다.
plt.plot(
  sr_one.index,
  sr_one.values
)

plt.show()
```

```python
plt.plot(
  sr_one
)

plt.show()
```

```python
# 한글 폰트 문제 해결 
# matplotlib은 한글 폰트를 지원하지 않음
# os정보
import platform

# font_manager : 폰트 관리 모듈
# rc : 폰트 변경 모듈
from matplotlib import font_manager, rc
# unicode 설정
plt.rcParams['axes.unicode_minus'] = False

if platform.system() == 'Darwin':
    rc('font', family='AppleGothic') # os가 macos
elif platform.system() == 'Windows':
    path = 'c:/Windows/Fonts/malgun.ttf' # os가 windows
    font_name = font_manager.FontProperties(fname=path).get_name()
    rc('font', family=font_name)
else:
    print("Unknown System")
```

```python
# 그림 사이즈 지정 (가로 : 14인치, 세로 : 5인지)
plt.figure(figsize=(14, 5))

# 데이터로 차트 그리기
plt.plot(
  sr_one
)

# 제목
plt.title('서울 -> 경기 인구 이동')  

# 축 이름 추가
plt.xlabel('기간')
plt.ylabel('이동 인구수')

# x축 눈금 라벨 회전하기
plt.xticks(rotation='vertical')  # 숫자 가능 

# 범례
plt.legend(
  labels = ['서울->경기'],
  loc='best'
)

plt.show()
```

```python
# 그림 사이즈 지정 (가로 : 14인치, 세로 : 5인지)
plt.figure(figsize=(14, 5))

# 데이터로 차트 그리기
plt.plot(
  sr_one
)

# 제목
plt.title('서울 -> 경기 인구 이동')  

# 축 이름 추가
plt.xlabel('기간')
plt.ylabel('이동 인구수')

# x축 눈금 라벨 회전하기
plt.xticks(rotation=45)  # 숫자 가능 

# y축 눈금 라벨 회전하기
plt.yticks(rotation=45)  # 숫자 가능 

# 범례
plt.legend(
  labels = ['서울->경기'],
  loc='best'
)

plt.show()
```

#### 4지역을 화면 분할하여 시각화 하기

```python
# 서울에서 충청남도, 경상북도, 강원도, 전라남도로 이동한 인구데이터

df_4 = df_seoul.loc[['충청남도','경상북도','강원도','전라남도']]
df_4
```

```python
# Style
plt.style.use('ggplot')

# 객체 
fig = plt.figure(figsize=(20, 10))
ax1 = fig.add_subplot(2, 2, 1)
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, 3)
ax4 = fig.add_subplot(2, 2, 4)

# 데이터 출력
ax1.plot(
  df_4.loc['충청남도',:],
  marker='o',
  markerfacecolor = 'green',
  markersize=10,
  color='olive',
  linewidth=2,
  label = '서울->충남'
)

ax2.plot(
  df_4.loc['경상북도',:],
  marker='o',
  markerfacecolor = 'blue',
  markersize=10,
  color='skyblue',
  linewidth=2,
  label = '서울->경북'
)

ax3.plot(
  df_4.loc['강원도',:],
  marker='o',
  markerfacecolor = 'red',
  markersize=10,
  color='magenta',
  linewidth=2,
  label = '서울->강원'
)

ax4.plot(
  df_4.loc['전라남도',:],
  marker='o',
  markerfacecolor = 'orange',
  markersize=10,
  color='yellow',
  linewidth=2,
  label = '서울->전남'
)

# 범례
ax1.legend(loc='best')
ax2.legend(loc='best')
ax3.legend(loc='best')
ax4.legend(loc='best')

# 제목
ax1.set_title('서울->충남 인구 이동')
ax2.set_title('서울->경북 인구 이동')
ax3.set_title('서울->강원 인구 이동')
ax4.set_title('서울->전남 인구 이동')

# 축눈금 라벨 지정 및 회전 
ax1.set_xticklabels(df_4.columns, rotation=45)
ax2.set_xticklabels(df_4.columns, rotation=45)
ax3.set_xticklabels(df_4.columns, rotation=45)
ax4.set_xticklabels(df_4.columns, rotation=45)

# 라벨 사이즈
ax1.tick_params(axis='x', labelsize=7)
ax2.tick_params(axis='x', labelsize=7)
ax3.tick_params(axis='x', labelsize=7)
ax4.tick_params(axis='x', labelsize=7)

# y축 범위
ax1.set_ylim(1000, 70000)
ax2.set_ylim(1000, 70000)
ax3.set_ylim(1000, 70000)
ax4.set_ylim(1000, 70000)

plt.show()
```

- 남북한 발전량 데이터셋을 사용
- 막대 그래프는 북한의 발전량을 표시
- 선그래프는 북한 발전량의 전년대비 증감율을 백분율로 나타낸다.

```python
df = pd.read_excel("./Data/남북한발전전력량.xlsx")
df
```

```python
# 북한 전력량만 선택 
df=df.loc[5:9]
df
```

```python
# 전력량 컬럼 삭제 
df.drop('전력량 (억㎾h)', axis='columns', inplace=True)
df
```

```python
# 발전 전력별을 index
df.set_index('발전 전력별', inplace=True)
df
```

```python
# 기준점을 발전 전력별로 변경
df = df.T
df.head()
```

```python
# 합계 -> 총발전량
df = df.rename(
  columns={'합계':'총발전량'}
)
df.head()
```

```python
# 총발전량 - 1년 이라는 컬럼
df['총발전량 - 1년'] = df['총발전량'].shift(1)
df.head()
```

```python
# 증감율 컬럼 생성
df['증감율'] = ((df['총발전량'] / df['총발전량 - 1년']) - 1) * 100
df.head()
```

# 2축 그래프 그리기

```python
ax1 = df[['수력', '화력']].plot(
  kind='bar',
  figsize=(20, 10),
  stacked=True
)

ax2 = ax1.twinx()

ax2.plot(
  df.index,
  df.증감율,
  ls='--',
  marker='o',
  color='green',
  label='전년대비 증감율(%)'
)

ax1.set_ylim(0, 500)
ax2.set_ylim(-50, 50)

ax1.set_xlabel('연도')
ax1.set_ylabel('발전량')
ax2.set_ylabel('전년대비 증감율')

plt.title('북한 전력 발전량')

ax1.legend(loc='upper left') # lower / right

plt.show()
```
