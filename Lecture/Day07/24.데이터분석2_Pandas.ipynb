```python
import numpy as np
import pandas as pd
```

```python
data = {
  "names" : ['kenny', 'kenny', 'kenny','charles','charles'],
  'year' :  [2016,       2017,    2018,     2016, 2017],
  "points" :[1.5,        1.77,     3.6,      2.4, 2.9]
}

df = pd.DataFrame(
  data,
  columns=['year','names','points','penalty'],
  index=['one','two','three','four','five']
)
df
```

```python
df.penalty = 0.5
df
```

```python
# range(0.1, 0.6, 0.1)
np.arange(0.1, 0.6, 0.1)
```

```python
df.penalty = np.arange(0.1, 0.6, 0.1)
df
```

```python
df['zeros'] = range(5)
df
```

```python
# Series
val = pd.Series(
  [-1.2, -1.5, -1.7],
  index=['two', 'four','five']
)

val
```

```python
df['debt'] = val
df
```

```python
# net_points = points - penalty
df['net_points'] = df.points - df.penalty
df
```

```python
df['high_points'] = df.net_points > 2
df
```

```python
# df에서 high_points를 삭제

del df['high_points']
df
```

```python
del df['zeros']
del df['net_points']
df
```
