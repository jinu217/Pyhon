### requests 
#### 응답코드   
1XX : 요청을 받았고 작업 진행중       
2XX : 사용자의 요청이 성공적으로 수행됨.      
3XX : 요청은 완료되었으나 리다이렉션이 필요.     
4XX : 사용자의 요청이 잘못됨.    
5XX : 서버에 오류가 발생함.    

```python
# request 응답 코드
import requests

URL = "https://www.naver.com"
response = requests.get(URL)
print(response.status_code)
```

```python
# 메인페이지 내용 가져오기
import requests

URL = "https://www.naver.com"
response = requests.get(URL)
print(response.text)
```

### 데이터 다운로드 하기

```python
import urllib.request as request

# 다운로드 파일 이름 결정
fileName = "mov01.jpg"

# URL과 저장경로 지정하기
url = "https://zeushahn.github.io/Test/images/" + fileName
saveName = "./Data/" + fileName

# 다운로드
request.urlretrieve(url, saveName)
print("OK!")
```

```python
# mov02.jpg, mov03.jpg. mov04.jpg

fileLists = ["mov02.jpg", "mov03.jpg", "mov04.jpg"]

for fileName in fileLists:
  url = "https://zeushahn.github.io/Test/images/" + fileName
  saveName = "./Data/" + fileName
  request.urlretrieve(url, saveName)
  print(f"{fileName} is downloaded.")
```
