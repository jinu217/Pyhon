### 서울시 유가 정보 분석

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# file name 합치기
from glob import glob 
```

```python
station_files = glob("./Data/oil/*.xls")
```

```python
# 파일 이름을 정렬
station_files = sorted(station_files)
station_files
```

```python
# xls를 읽기 위해
!pip install xlrd
```

```python
tmp_raw = []
for file_name in station_files:
  tmp = pd.read_excel(file_name, header=2)
  tmp_raw.append(tmp)

station_raw = pd.concat(tmp_raw)
```

```python
len(station_raw)
```

```python
station_raw.head()
```

```python
station_raw.info()
```

```python
# 필요항목 발췌
stations = pd.DataFrame(
  {
    'Oil_store' : station_raw['상호'],
    '주소' : station_raw['주소'],
    '가격' : station_raw['휘발유'],
    '셀프' : station_raw['셀프여부'],
    '상표' : station_raw['상표']
  }
)
stations.head()
```

```python
# 주소 컬럼에 서울특별시로 입력되어 있는 부분 확인
[region.split()[0] for region in stations['주소']]
```

```python
# 주소 컬럼에 주소의 앞부분의 unique 확인 
set([region.split()[0] for region in stations['주소']])
```
